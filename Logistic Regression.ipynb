```python
def perceptron(x,y):
    weight=np.ones(x.shape[1])
    learn=0.1
    for i in range(1000):
        select=np.random.randint(x.shape[0])
        y_hat=sigmoid(np.dot(weight,x[select]))
        weight=weight + learn*(y[select]-y_hat)*x[select]
    return weight
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def gd(x,y):
    weight=np.ones(x.shape[1])
    learn=0.1
    for i in range(1000):
        y_hat=sigmoid(np.dot(x,weight))
        weight=weight+learn*np.dot((y-y_hat),x)*1/x.shape[0]
    return weight
```
